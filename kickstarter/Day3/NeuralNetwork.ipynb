```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Model
from sklearn.neural_network import MLPClassifier
# Data process
from sklearn.model_selection import KFold, train_test_split
# Metrics
from sklearn.metrics import log_loss, accuracy_score, confusion_matrix
```

```python
# データ読み込み
df = pd.read_csv('../data/day2_onehot.csv')
df.head()
```

```python
# 標準化
from sklearn.preprocessing import StandardScaler
```

```python
stdsc = StandardScaler()
x_df_std = df.drop(['days', 'goal'], axis=1)
x_df_std['std_days'] = stdsc.fit_transform(df['days'].values.reshape(-1,1))
x_df_std['std_goal'] = stdsc.fit_transform(df['goal'].values.reshape(-1,1))
x_df_std.head()
```

```python
X = x_df_std.iloc[:,1:].values
y = df["state"].values * 1

test_size = 0.2        # 全データのうち、何%をテストデータにするか（今回は20%に設定）
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=1234) # ホールドアウト法を実行（テストデータはランダム選択）
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

clf = MLPClassifier(hidden_layer_sizes=(100,100,), activation='relu', solver='adam', batch_size=50)
clf = clf.fit(X_train, y_train)
print("score=", clf.score(X_test, y_test))
print(clf.predict(X_test)) #予測したい場合
```
