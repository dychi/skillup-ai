#  活性化関数

```python
import numpy as np
import matplotlib.pyplot as plt
```

## ステップ関数

```python
def step(x):
    return np.array(x>0).astype(np.int)
    
x = np.arange(-5,5,0.1)
y = step(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-0.1,1.1])
plt.show()
```

## シグモイド関数

```python
def sigmoid(x):
    return 1 / (1+ np.exp(-x))

x = np.arange(-5,5,0.1)
y = sigmoid(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-0.1,1.1])
plt.show()
```

## tanh関数

```python
def tanh(x):
    return np.tanh(x)

x = np.arange(-5,5,0.1)
y = tanh(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-2,2])
plt.show()
```

## ReLU関数
- $\mathrm{ReLU(x)} = \max{(x, 0)}$
- $x$が0以上ならば$x$を、0未満なら0と出力する

```python
def relu(x):
    """
    この関数を編集する
    """
    x = np.maximum(0, x)
    return x

x = np.arange(-5,5,0.1)
y = relu(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-5.0,5])
plt.show()
```
