#  活性化関数

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 恒等写像関数
$h(x)=x$

```python
def linear(x):
    return x
    
x = np.arange(-5,5,0.1)
y = linear(x) 
plt.figure(figsize=(3.5,3))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-5,5])
plt.show()
```

## ステップ関数

```python
# ヒント
a = np.array([-1,0,1])
print(a > 0)
np.array([True, False]).astype(np.int)
```

```python
def step(x):
    return (x > 0) * 1 
    # return np.array(x > 0).astype(np.int)
    
x = np.arange(-5,5,0.1)
y = step(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-0.1,1.1])
plt.show()
```

## シグモイド関数

$\displaystyle{h(x)=\frac{1}{(1+\exp(-x))}}$

```python
# ヒント
np.exp(1)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5,5,0.1)
y = sigmoid(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-0.1,1.1])
plt.show()
```

## tanh関数

$\displaystyle h(x) = \frac{\sinh(x)}{\cosh(x)}=\frac{e^x-e^{-x}}{e^x+e^{-x}}$  

```python
# ヒント
np.sinh(1),np.cosh(1),np.tanh(1)
```

```python
def tanh(x):
    return np.tanh(x)

x = np.arange(-5,5,0.1)
y = tanh(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-2,2])
plt.show()
```

## ReLU関数

$h(x) = x (x>=0)$  
$h(x) = 0 (x<0)$

```python
# ヒント
np.maximum(1, np.array([-1, 0, 1, 2]))
```

```python
def relu(x):
    return np.maximum(x, 0)

x = np.arange(-5,5,0.1)
y = relu(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-5.0,5])
plt.show()
```

## LeakyReLU
$h(x) = x (x>=0)$  
$h(x) = ax (x<0)$

```python
def leakyrelu(x):
    return np.maximum(x, 0.1 * x) 

x = np.arange(-5,5,0.1)
y = leakyrelu(x) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-5.0,5])
plt.show()
```

## Softplus
$h(x)=\log(1+\exp(x))$

```python
def softplus(x):
    return np.log(1 + np.exp(x))

x = np.arange(-5,5,0.1)
y = softplus(x) 
plt.figure(figsize=(3.5,3))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-5.0,5])
plt.show()
```

## Hardtanh
$h(x) = x (-1<=x, x<=1)$  
$h(x) = 1 (x>1)$  
$h(x) = -1 (x<-1)$

```python
def hardtanh(x):
    return np.maximum(-1, np.minimum(1, x))

x = np.arange(-5,5,0.1)
y = hardtanh(x) 
plt.figure(figsize=(3.5,3))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-5.0,5])
plt.show()
```
